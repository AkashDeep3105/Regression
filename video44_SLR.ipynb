# DATA PREPROCESSING

IMPORT MODULE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

IMPORT DATASET

```python
dataset=pd.read_csv("C:\\Users\\ad310\\Desktop\\MACHINE LEARNING\\drive-download-20241214T151724Z-001\\Part 2 - Regression\\Section 4 - Simple Linear Regression\\Python\\Salary_Data.csv")
x=dataset.iloc[ : , :-1].values
y=dataset.iloc[ : ,-1].values
```

SPLIT DATASET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# TRAINING THE MODEL

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

# PREDICTING RESULT

```python
y_pred=regressor.predict(x_test)
```

# Visualising the Training set results

TRAINING SET

```python
x_train_pred=regressor.predict(x_train)
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,x_train_pred,color='blue')
plt.title('salary vs experience(test)')
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,x_train_pred,color='blue')
plt.title('salary vs experience(test)')
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```
